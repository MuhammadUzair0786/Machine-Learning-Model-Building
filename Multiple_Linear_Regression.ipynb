# Multiple Linear Regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\Muhammad Uzair\Downloads\Datasets\multiple_linear_regression_dataset.csv')
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(dataset, kind='reg', diag_kind='kde')
plt.show()
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
input_data = dataset.iloc[:, :-1]
output_data = dataset['salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
# Predict the salary for the first row in X_test
model.predict([X_test.iloc[0].values])
```
