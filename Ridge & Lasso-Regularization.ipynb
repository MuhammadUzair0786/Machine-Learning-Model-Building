```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in dataset.select_dtypes(include=['object']).columns:
    dataset[i] = label_encoder.fit_transform(dataset[i])
    
dataset.head()
```

```python
# Splitting the dataset into features and target variable
x = dataset.iloc[:,:-1]
y = dataset['SalePrice']
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(f"Train Score Linear Regression : {model.score(x_train, y_train)*100:.2f}%")
print(f"Test Score Linear Regression : {model.score(x_test, y_test)*100:.2f}%")
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
poly_model = LinearRegression()
poly_model.fit(x_poly, y_train)
```

```python
print(f"Train Score Polynomial_Regression : {poly_model.score(x_poly, y_train)*100:.2f}%")
print(f"Test Score Polynomial_Regression : {poly_model.score(poly.transform(x_test), y_test)*100:.2f}%")
```

# Lasso (L1) Reglurization

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1, max_iter=10000, random_state=42)
# Fitting the Lasso model
lasso.fit(x_train, y_train)
```

```python
print(f"Train Score Lasso Regression : {lasso.score(x_train, y_train)*100:.2f}%")
print(f"Test Score Lasso Regression : {lasso.score(x_test, y_test)*100:.2f}%")
```

# Ridge (L2) Regularization

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1, max_iter=10000, random_state=42)
# Fitting the Ridge model
ridge.fit(x_train, y_train)

```

```python
print(f"Train Score Ridge Regression : {ridge.score(x_train, y_train)*100:.2f}%")
print(f"Test Score Ridge Regression : {ridge.score(x_test, y_test)*100:.2f}%")
```
