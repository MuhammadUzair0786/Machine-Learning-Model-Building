# Simple Linear Regression Model

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Muhammad Uzair\Downloads\Datasets\Salary_dataset.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=dataset)
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
input_data = dataset[['YearsExperience']]
output_data = dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=30)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
dataset.head(5)
```

```python
model.predict(pd.DataFrame([[2.3]], columns=['YearsExperience']))
```

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=dataset)
plt.plot(input_data['YearsExperience'], model.predict(input_data), color='red', linewidth=2)
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['Data Points','Predicted Line'])
plt.grid()
plt.show()
```
